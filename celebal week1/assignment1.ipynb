Import Required Libraries

```python
import pandas as pd

print("Pandas library imported successfully!")
```

Question 1: Load a CSV Dataset into a Pandas DataFrame

```python
df = pd.read_csv("C:\\Users\\Acer\\Desktop\\Combined_dataset.csv")

print("Dataset loaded successfully!")
```

Question 2: Explore Data (head/tail, shape, columns, data types)

```python
# 2. Explore data
print("--- First 5 Rows (Head) ---")
display(df.head())

print("\n--- Last 5 Rows (Tail) ---")
display(df.tail())

print("\n--- Dataset Shape (Rows, Columns) ---")
print(df.shape)

print("\n--- Column Names ---")
print(df.columns.tolist())

print("\n--- Data Types ---")
print(df.dtypes)
```

Question 3: Handle Missing Values (identify, fill/drop)

```python

print("--- Missing Values Count Before Cleaning ---")
print(df.isnull().sum())

if 'price' in df.columns and 'quantity' in df.columns:
    df.dropna(subset=['price', 'quantity'], inplace=True)


df.fillna('Unknown', inplace=True)

print("\n--- Missing Values Count After Cleaning ---")
print(df.isnull().sum())
```

Question 4: Perform Basic Operations (filter rows, select columns)

```python
sales_data = df
if 'ratings_count' in sales_data.columns:

    sales_data['ratings_count'] = pd.to_numeric(sales_data['ratings_count'], errors='coerce')
    sales_data = sales_data[sales_data['ratings_count'] > 0]

columns_to_select = ['title', 'initial_price', 'final_price', 'ratings_count']

existing_columns = [col for col in columns_to_select if col in sales_data.columns]

if existing_columns:
    selected_columns = sales_data[existing_columns]
    
    print("\n--- Subset of Selected Columns ---")
    display(selected_columns.head())
```

Question 5: Remove Duplicates

```python
# 5. Remove duplicates
print(f"Shape before dropping duplicates: {df.shape}")

df.drop_duplicates(inplace=True)

print(f"Shape after dropping duplicates: {df.shape}")
```

Question 6: Create a Derived Column (total_amount = price * quantity)

```python
sales_data = df

col_price = 'initial_price'
col_qty = 'ratings_count'

if col_price in sales_data.columns and col_qty in sales_data.columns:
    sales_data[col_price] = pd.to_numeric(sales_data[col_price], errors='coerce')
    sales_data[col_qty] = pd.to_numeric(sales_data[col_qty], errors='coerce')
    
    sales_data['estimated_revenue'] = sales_data[col_price] * sales_data[col_qty]
    
    print("\n--- DataFrame with Derived 'estimated_revenue' Column ---")
    display(sales_data[[col_price, col_qty, 'estimated_revenue']].head())
    
else:
    print(f"Warning: Missing required columns. Could not find '{col_price}' or '{col_qty}'.")
```

Question 7: Save the Cleaned Dataset as a New CSV File

```python
output_filename = 'cleaned_dataset.csv'
df.to_csv(output_filename, index=False)

print(f"Cleaned dataset successfully saved as '{output_filename}'!")
```
